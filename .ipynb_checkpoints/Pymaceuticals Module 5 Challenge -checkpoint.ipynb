```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from pathlib import Path
import numpy as np

# Study data files
mouse_metadata_path ='data/mouse_metadata.csv'
study_results_path = 'data/Study_results.csv'

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
combined_mouse_study_df=pd.merge(study_results,mouse_metadata,how="left",on="Mouse ID")
combined_mouse_study_df

# Display the data table for preview
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
grouped_combined_mouse_study_df = combined_mouse_study_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].agg(['mean', 'median', 'var', 'std', 'sem']).reset_index()

# Rename the columns of the new dataframe for better understanding
grouped_combined_mouse_study_df.columns = ['Regimens', 'Mean', 'Median', 'Variance', 'StandardDeviation', 'SEM']

mean_value = np.mean(combined_mouse_study_df['Tumor Volume (mm3)'])
median_value = np.median(combined_mouse_study_df['Tumor Volume (mm3)'])
variance_value = np.var(combined_mouse_study_df['Tumor Volume (mm3)'])
std_deviation_value = np.std(combined_mouse_study_df['Tumor Volume (mm3)'],ddof=1)
sem_value = std_deviation_value / np.sqrt(len(combined_mouse_study_df['Tumor Volume (mm3)']))

stats_dict_TV = {
    "Mean": [mean_value],
    "Median": [median_value],
    "Variance": [variance_value],
    "Standard Deviation": [std_deviation_value],
    "SEM": [sem_value]}

# Create the dataframe
stats_summary_TV = pd.DataFrame(stats_dict_TV)
stats_summary_TV

# Assemble the resulting series into a single summary DataFrame.
grouped_combined_mouse_study_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
max_tumor = clean_combined_df.groupby(["Mouse ID"])['Timepoint'].max()
max_tumor = max_tumor.reset_index()

merged_data = max_tumor.merge(clean_combined_df,on=['Mouse ID','Timepoint'],how="left")
merged_data

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
```

```python

# Put treatments into a list for for loop (and later for plot labels)
treatments=['Capomulin','Ramicane','Infubinol','Ceftamin']


# Create empty list to fill with tumor vol data (for plotting)
tumor_list=[]

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for drug in treatments:
    
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    drug_data = merged_data[merged_data["Drug Regimen"] == drug]["Tumor Volume (mm3)"]
    
    # add subset 
    tumor_list.append(drug_data)
    
    # Determine outliers using upper and lower bounds
    q1 = drug_data.quantile(0.25)
    q3 = drug_data.quantile(0.75)
    iqr = q3-q1 #Interquartile range
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr
    df_out = drug_data.loc[(drug_data < fence_low) | (drug_data > fence_high)]
    print(f"{drug}'s potential outliers: {df_out}")
    
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
flierprops = dict(marker='o', markerfacecolor='r', markersize=10,
linestyle='none', markeredgecolor='black')
plt.boxplot(tumor_list,labels=treatments,flierprops=flierprops)
plt.ylabel("Final Tumor Volume (mm3)")
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
cap_df=clean_combined_df.loc[clean_combined_df["Drug Regimen"]=="Capomulin"]
cap_agg_df=cap_df.groupby("Mouse ID")[["Weight (g)","Tumor Volume (mm3)"]].mean()
x=cap_agg_df["Weight (g)"]
y=cap_agg_df["Tumor Volume (mm3)"]
plt.scatter(x,y)
plt.ylabel("Average Tumor Volume (mm3)")
plt.xlabel("Weight (g)")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen

cap_df=clean_combined_df.loc[clean_combined_df["Drug Regimen"]=="Capomulin"]
cap_agg_df=cap_df.groupby("Mouse ID")[["Weight (g)","Tumor Volume (mm3)"]].mean()
x=cap_agg_df["Weight (g)"]
y=cap_agg_df["Tumor Volume (mm3)"]
plt.scatter(x,y)
slope, intercept, r, p, se = st.linregress(x, y)
y_values=slope*x+intercept
plt.plot(x,y_values,color="r")
print(f"The correlation between mouse weight and the average tumor volume is {round(r,2)}")
plt.ylabel("Average Tumor Volume (mm3)")
plt.xlabel("Weight (g)")
plt.show()
```
